```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10,10,20) #The code uses NumPy's linspace function to generate an array of 20 evenly spaced numbers between -10 and 10
```

```python
x #displays the code above
```

```python
m = 50
b = 17
```

```python
y = m * x + b #set's "y" to be equal to the slope intercept formula
```

```python
plt.plot(x,y,'.') #plot's the x and y values with "." so that way the values are displayed as dots
```

```python
r = 100*np.random.randn(20,) #code makes an array "r" containing 20 random numbers, with a mean of 0 and a standard deviation of 100
```

```python
r #displays the code above
```

```python
plt.plot(r,'.') #plost r
```

```python
y2 = m * x + b + r
```

```python
plt.plot(x,y2,'.')
```

```python
m_guess = 1
b_guess = 1 #Define the guesses for the slope (m_guess) and y-intercept (b_guess)
y_guess = m_guess * x + b_guess #Calculate y_guess with the guessed slope * x and y-intercept

plt.plot(x,y_guess,'r.'); #plots the x and y_guess values with r. meaning it will show as red dots on the graph
plt.plot(x,y2,'g.'); #plots the x and y2 with g. meaning it will show as green dots on the graph.
```

```python
for m_guess in np.linspace(-100,100,100): #loop over a range of values for m_guess(-100 starting point, 100 end point, 100 values between the points)

        y_guess = m_guess * x #defines y_guess as multiplying the guess slope with x

        plt.plot(x,y_guess,'r.'); #plots x with y_guess as red dots
        plt.plot(x,y2,'g.'); #plots x and y2 as green dots
        plt.show() #Display the plot

        loss = np.sum((y_guess - y)**2) #calculate the loss between y_guess and y2 and squaring the difference with **2

        print(loss,'M = ',m_guess) #print the loss along with the value of m_guess
```

```python
best_loss = 100000000000 #large value to start
best_m = 0
best_b = 0 #best loss values for m and b

for m_guess in np.linspace(-100,100,100): #loop over the range of values in the() for m_guess
    for b_guess in np.linspace(-100,100,100): #same as the code above except for b_guess

        y_guess = m_guess * x + b_guess

        loss = np.sum((y_guess - y)**2) #calculate difference with y_guess and y then squaring it

        if loss < best_loss: #check if loss is less than best_loss
#if so then update best_loss and best_m and b
            best_loss = loss
            best_m = m_guess
            best_b = b_guess
```

```python
#plot the line
y_guess = best_m * x + best_b #slope formula with best_m and b

plt.plot(x,y_guess,'r.'); #plots the red dots
plt.plot(x,y2,'g.'); #plots the green dots
plt.show() #display the plot

print(best_loss,best_m,best_b) #print the loss, slope(best_m) and intercept(best_b)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
def get_database(url,sheet_name): #Define a function named get_database that takes the URL and sheet_name
    url_head = "https://docs.google.com/spreadsheets/d/"
    url_foot = "/gviz/tq?tqx=out:csv&sheet="
    url_body = url.split('/')[5]
    # sheet_name = "1"
    url_csv = url_head + url_body + url_foot + sheet_name
    df = pd.read_csv(url_csv, sep=',', skiprows=0)
    return df
```
